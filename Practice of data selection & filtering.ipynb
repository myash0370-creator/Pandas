## 🟢 Section 1: Column Selection (df[col])

#### Q1. Select only the age column

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'name': ['Amit', 'Riya', 'John', 'Sara', 'Raj'],
    'age': [20, 17, 22, 19, 18],
    'marks': [85, 90, 45, 70, 30]
})
```

```python
df['age']
```

#### Q2. Select both name and marks columns

```python
df[['name', 'age']]
```

## 🟡 Section 2: .loc[]

#### Q3. Select the row with index 2 using .loc[]

```python
df.loc[1]
```

#### Q4. Select rows from index 1 to 3 and columns name and marks

```python
df.loc[0:2]
```

#### Q5. Select the marks column using .loc[] (all rows) 

```python
df.loc[: , 'marks']
```

## 🟠 Section 3: .iloc[]

#### Q6. Select the 3rd row using .iloc[]

```python
df.iloc[2]
```

#### Q7. Select the first 4 rows

```python
df.iloc[0:4]
```

#### Q8. Select the value at row index 2 and column index 2

```python
df.iloc[2 , 2]
```

## 🔵 Section 4: .at[] and .iat[]

#### Q9. Get the marks value at index 3 using .at[]

```python
df.at[3 , 'marks']
```

#### Q10. Get the same value using .iat[]

```python
df.iat[3 , 2]
```

#### Q11. Get the name at index 0 using .at[]

```python
df.at[0 , 'name']
```

## 🔴 Section 5: Filtering with Conditions

#### Q12. Filter students with marks greater than 80

```python
df[df['marks'] > 80]
```

#### Q13. Filter students where age ≥ 18 AND marks > 50

```python
df[(df['age'] >= 18 ) & (df['marks'] >50)]
```

#### Q14. Filter students where marks < 50 OR age < 18

```python
df[(df['marks'] < 50 ) | (df['age'] < 18)]
```

## 🟣 Section 6: .query()

#### Q15. Filter students with marks > 70 using .query()

```python
df.query("marks > 70")
```

#### Q16. Filter students where age > 18 AND marks < 80 using .query()

```python
df.query("age > 18 & marks <80")
```

#### Q17. Using .query(), return only name and marks where marks > 60

```python
df.query("marks > 60") [['name' , 'marks']]
```

## 🧠 Final Challenge


#### Q18. In a single line:

#### Filter students with marks ≥ 50
#### age < 22
#### Return only the name column
#### (using .query() + selection)

```python
df.query("marks >= 50 & age <22") ['name']
```
