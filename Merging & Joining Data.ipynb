```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpID" : [1,2,3],
    "Name" : ["Yash" ,"Ashish" , "Suraj"],
    "DeptID": [10,20,30]
})
```

```python
employees
```

```python
departments = pd.DataFrame({
    "DeptID" : [10,20,40],
    "DeptName" : ["HR", "Engineering" , "Marketing"]
})
```

## Merge Like SQL: pd.merge()

```python
pd.merge(employees , departments , on="DeptID") #Inner Join
```

```python
pd.merge(employees , departments , on="DeptID", how="left") #Left join 
```

```python
pd.merge(employees, departments, on="DeptID", how="right") # Right Join
```

```python
pd.merge(employees, departments, on="DeptID", how="outer") # Right Join
```

## Concatenating data frames

```python
df1 = pd.DataFrame({"Name": ["Alice", "Bob"], "score": [423, 634]})
df2 = pd.DataFrame({"Name": ["Charlie", "David"], "age": [45, 78]})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1 , df2])
```

```python
pd.concat([df1 , df2], axis=1)
```
