```python
import pandas as pd
```

```python
data = {
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance'],
    'Salary': [50000, 60000, 40000, 45000, 70000],
    'Age': [25, 30, 35, 28, 40]
}
df = pd.DataFrame(data)
```

```python
df.groupby('Department')['Salary'].mean() # Department wise average salary
```

```python
df.groupby('Department')['Salary'].sum() # every Department Total Salary
```

```python
df.groupby('Department').agg({        #.groupby() + .agg() use both toh calaculate avaerage salary and max age
    'Salary' : 'mean',
    'Age' : 'max'
})
```

```python
df["Avg Salary"] = df.groupby('Department')['Salary'].transform('mean') #Add ner coloumn average salary
df
```

```python
df[df['Avg Salary']>50000]  # on;y those departments whose average salary has more than 50000.
```
