## 🔹Sorting & Ranking

```python
df = pd.DataFrame({
    "Name": ["A", "B", "C", "D"],
    "Marks": [70, 50, 90, 60]
})
```

#### Q1: Sort the DataFrame by Marks in descending order.

```python
df.sort_values("Marks" , ascending=False)
```

#### Q2: Create a new column "Rank" based on Marks.

```python
df["Rank"] = df["Marks"].rank(ascending=False)
df
```

## Renaming Columns & Index



```python
df = pd.DataFrame({
    "name": ["A", "B"],
    "marks": [50, 60]
})
```

#### Q1: Rename "name" → "Name" and "marks" → "Marks".

```python
df.rename(columns={"name" : "Name" , "marks" : "Marks"} , inplace = True)
df
```

#### Q2: Change index to ["a", "b"]

```python
df.rename(index={0:"a", 1: "b"}, inplace=True)
df
```
