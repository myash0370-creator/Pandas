## 🔹 1. Missing Values

#### Q1: Find missing values in each column and fill missing values in "Age" with the mean.

```python
import pandas as pd

data = {
    "Name": ["Amit", "Baby", "Cat", "Dog"],
    "Age": [20, None, 22, None],
    "City": ["Delhi", "Mumbai", None, "Chennai"]
}
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].mean())
```

#### Q2: Drop all rows where any value is missing.

```python
df.dropna()
```

## 🔹 3. Type Conversion

## 🔹 4. apply()

#### Q1: Create "Result": Fail/Pass.

```python
df = pd.DataFrame({
    "Marks": [40, 55, 70, 90]
})
```

```python
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x > 50 else "Fail")
df
```

#### Q2: Create "Square" column.

```python
df["Result"] = df["Marks"].apply(lambda x: x*x)
df
```

## 🔹 5. map()

```python
df = pd.DataFrame({
    "Grade": ["A", "B", "C", "A"],
    "Status": ["Yes", "No", "Yes", "No"]
})
```

#### Q1: Map grades to words.

```python
grade_map = {"A" : "Excellent" , "B" : "Good" , "C" : "Average"}
df["Grade"] = df["Grade"].map(grade_map)
df
```

#### Q2: Convert Yes/No to True/False.

```python
status_map = {"Yes" : "True", "No" : "False"}
df["status"] = df["Status"].map(status_map)
df
```

## 🔹 6. replace()

#### Q1: Replace Delhi → NCR.

```python
df = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai"],
    "Grade": ["A", "B", "C", "A"]
})
```

```python
df["City"].replace({"Delhi" : "Delhi NCR"})
```

#### Q2: Replace grades with numbers.

```python
df["Grade"].replace({"A" : 1 , "B" :2 , "C" : 3})
```
